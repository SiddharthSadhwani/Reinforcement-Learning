```python
def one_sim(k,epsilon,var,time_steps):        # 1 epsilon-greedy simulation
    avg_reward=[0]*time_steps
    tot_reward=0
    optimal_action=[0]*time_steps
    error=[0]*time_steps
    q_star=np.random.normal(loc=0,scale=1,size=(10))    # q* with mean=0 and variance 1
    optimal_arm=np.argmax(q_star)
    opt_selections=0
    num_selections=[0]*10
    q_t=[0]*10
    for i in range(time_steps):
        p=random.random()
        if p<(1-epsilon):        # exploitation
            arm=np.argmax(q_t)
        else:                    # exploration
            arm=random.randint(0,9)
        num_selections[arm]+=1
        if optimal_arm==arm:
            optimal_action[i]=1
        reward_t=np.random.normal(loc=q_star[arm],scale=var)        # generating reward with mean q* and variance 1
        error[i]=abs(reward_t-q_t[arm])
        q_t[arm]=(q_t[arm])*((num_selections[arm]-1)/(num_selections[arm]))+reward_t/(num_selections[arm])
        avg_reward[i]=reward_t
    return avg_reward,optimal_action,error
```

```python
def multi_sim(k,epsilon,var,exp,time_steps):        # 2000 simulations
    avg_reward=[0]*time_steps
    optimal_action=[0]*time_steps
    error=[0]*time_steps
    for i in range(exp):
        a,b,c=one_sim(k,epsilon,var,time_steps)
        for j in range(time_steps):
            avg_reward[j]+=a[j]/exp
            optimal_action[j]+=(b[j]/exp)*100
            error[j]+=c[j]/exp
    return avg_reward,optimal_action,error
```

```python
def one_sim_t(k,var,time_steps):        # 1 epsilon greedy simulation but with epsilon = 1/t
    avg_reward=[0]*time_steps
    tot_reward=0
    optimal_action=[0]*time_steps
    error=[0]*time_steps
    q_star=np.random.normal(loc=0,scale=1,size=(10))
    optimal_arm=np.argmax(q_star)
    opt_selections=0
    num_selections=[0]*10
    q_t=[0]*10
    for i in range(time_steps):
        epsilon=1/(i+1)            # epsilon = 1/t
        p=random.random()
        if p<(1-epsilon):
            arm=np.argmax(q_t)
        else:
            arm=random.randint(0,9)
        num_selections[arm]+=1
        if optimal_arm==arm:
            optimal_action[i]=1
        reward_t=np.random.normal(loc=q_star[arm],scale=var)
        error[i]=abs(reward_t-q_t[arm])
        q_t[arm]=(q_t[arm])*((num_selections[arm]-1)/(num_selections[arm]))+reward_t/(num_selections[arm])
        avg_reward[i]=reward_t
    return avg_reward,optimal_action,error
```

```python
def multi_sim_t(k,var,exp,time_steps):        # 2000 simulations for epsilon = 1/t
    avg_reward=[0]*time_steps
    optimal_action=[0]*time_steps
    error=[0]*time_steps
    for i in range(exp):
        a,b,c=one_sim_t(k,var,time_steps)
        for j in range(time_steps):
            avg_reward[j]+=a[j]/exp
            optimal_action[j]+=(b[j]/exp)*100
            error[j]+=c[j]/exp
    return avg_reward,optimal_action,error
```

```python
def one_sim_sum(k,epsilon,var,time_steps):  # 1 epsilon greedy simulation for cumulative rewards
    avg_reward=[0]*time_steps
    tot_reward=0
    cumulative_sum=[0]*time_steps
    optimal_action=[0]*time_steps
    error=[0]*time_steps
    q_star=np.random.normal(loc=0,scale=1,size=(10))
    optimal_arm=np.argmax(q_star)
    opt_selections=0
    num_selections=[0]*10
    q_t=[0]*10
    for i in range(time_steps):
        p=random.random()
        if p<(1-epsilon):
            arm=np.argmax(q_t)
        else:
            arm=random.randint(0,9)
        num_selections[arm]+=1
        if optimal_arm==arm:
            optimal_action[i]=1
        reward_t=np.random.normal(loc=q_star[arm],scale=var)
        error[i]=abs(reward_t-q_t[arm])
        q_t[arm]=(q_t[arm])*((num_selections[arm]-1)/(num_selections[arm]))+reward_t/(num_selections[arm])
        if i==0:
            cumulative_sum[i]=reward_t
        else:
            cumulative_sum[i]+=cumulative_sum[i-1]+reward_t       # updating cumulative reward for 1 step
        avg_reward[i]=reward_t
    return avg_reward,optimal_action,error,cumulative_sum
```

```python
def multi_sim_sum(k,epsilon,var,exp,time_steps):        # 2000 simulations
    avg_reward=[0]*time_steps
    optimal_action=[0]*time_steps
    cumulative_sum=[0]*time_steps
    error=[0]*time_steps
    for i in range(exp):
        a,b,c,d=one_sim_sum(k,epsilon,var,time_steps)
        for j in range(time_steps):
            avg_reward[j]+=a[j]/exp
            optimal_action[j]+=(b[j]/exp)*100
            error[j]+=c[j]/exp
            cumulative_sum[j]+=d[j]/exp
    return avg_reward,optimal_action,error,cumulative_sum
```

```python
def one_sim_ns1(k,epsilon,var,time_steps):        # 1 simulation for non stationary condition using sample average
    avg_reward=[0]*time_steps
    tot_reward=0
    optimal_action=[0]*time_steps
    error=[0]*time_steps
    q_star=[np.random.normal(loc=0,scale=1)]*10
    opt_selections=0
    num_selections=[0]*10
    q_t=[0]*10
    for i in range(time_steps):
        optimal_arm=np.argmax(q_star)
        p=random.random()
        if p<(1-epsilon):
            arm=np.argmax(q_t)
        else:
            arm=random.randint(0,9)
        num_selections[arm]+=1
        if optimal_arm==arm:
            optimal_action[i]=1
        reward_t=np.random.normal(loc=q_star[arm],scale=var)
        error[i]=abs(reward_t-q_t[arm])
        q_t[arm]=(q_t[arm])*((num_selections[arm]-1)/(num_selections[arm]))+reward_t/(num_selections[arm])        # sample average
        avg_reward[i]=reward_t
        for j in range(k):
            q_star[j]+=np.random.normal(loc=0,scale=0.01)        # increment in q* acc. to normal dist. mean 0 and sd=0.01
    return avg_reward,optimal_action,error
```

```python
def one_sim_ns2(k,epsilon,var,alpha,time_steps):        # # 1 simulation for non stationary condition using constant step size
    avg_reward=[0]*time_steps
    tot_reward=0
    optimal_action=[0]*time_steps
    error=[0]*time_steps
    q_star=[np.random.normal(loc=0,scale=1)]*10
    opt_selections=0
    q_t=[0]*10
    for i in range(time_steps):
        optimal_arm=np.argmax(q_star)
        p=random.random()
        if p<(1-epsilon):
            arm=np.argmax(q_t)
        else:
            arm=random.randint(0,9)
        if optimal_arm==arm:
            optimal_action[i]=1
        reward_t=np.random.normal(loc=q_star[arm],scale=var)
        error[i]=abs(reward_t-q_t[arm])
        q_t[arm]=(q_t[arm])+alpha*(reward_t-q_t[arm])        # constant step size update
        avg_reward[i]=reward_t
        for j in range(k):
            q_star[j]+=np.random.normal(loc=0,scale=0.01)        # increment in q* acc. to normal dist. mean 0 and sd=0.01
    return avg_reward,optimal_action,error
```

```python
def one_sim_ucb(k,epsilon,var,c,time_steps): # 1 simluation ,ucb with c=2
    avg_reward=[0]*time_steps
    tot_reward=0
    optimal_action=[0]*time_steps
    error=[0]*time_steps
    q_star=np.random.normal(loc=0,scale=1,size=(10))
    optimal_arm=np.argmax(q_star)
    opt_selections=0
    num_selections=[0]*10
    ucb=[0]*10
    q_t=[0]*10
    for i in range(time_steps):
        p=random.random()
        maximizing_action=-1
        for j in range(k):
            if num_selections[j]==0:        # if n(a) = 0, select arm as maximizing action
                maximizing_action=j
                break
            else:
                ucb[j]=q_t[j]+(c*math.sqrt(math.log(i+1)/num_selections[j]))        # ucb update step 
        if maximizing_action!=-1:
            arm=maximizing_action
        else:
            arm=np.argmax(ucb)
        num_selections[arm]+=1
        if optimal_arm==arm:
            optimal_action[i]=1
        reward_t=np.random.normal(loc=q_star[arm],scale=var)
        error[i]=abs(reward_t-q_t[arm])
        q_t[arm]=(q_t[arm])*((num_selections[arm]-1)/(num_selections[arm]))+reward_t/(num_selections[arm])
        avg_reward[i]=reward_t
    return avg_reward,optimal_action,error
```

```python
def multi_sim_ucb(k,epsilon,var,exp,c,time_steps):        # 2000 ucb simulations
    avg_reward=[0]*time_steps
    optimal_action=[0]*time_steps
    error=[0]*time_steps
    for i in range(exp):
        x,y,z=one_sim_ucb(k,epsilon,var,c,time_steps)
        for j in range(time_steps):
            avg_reward[j]+=x[j]/exp
            optimal_action[j]+=(y[j]/exp)*100
            error[j]+=z[j]/exp
    return avg_reward,optimal_action,error
```

```python
def preferance_selected(A_t,H_t,alpha,R_t,R_avg,pie_t):
    H_t1=H_t+alpha*(R_t-R_avg)*(1-pie_t)
    return H_t1
```

```python
def preferance_not_selected(A_t,H_t,alpha,R_t,R_avg,pie_t):
    H_t1=H_t-alpha*(R_t-R_avg)*(pie_t)
    return H_t1
```

```python
def softmax(A_t,H_t):
    sum_H=0
    for i in H_t:
        sum_H+=math.exp(i)
    return math.exp(H_t[A_t])/sum_H
```

```python
def one_sim_gb(k,alpha,var,time_steps):        # gradient bandit , 1 simulation with baseline
    avg_reward=[0]*time_steps
    total_reward=0
    optimal_action=[0]*time_steps
    q_star=np.random.normal(loc=4,scale=1,size=(k))
    optimal_arm=np.argmax(q_star)
    opt_selections=0
    num_selections=[0]*k
    R_avg=0
    H_t=[0]*k
    p_t=[0.1]*k
    for i in range(time_steps):
        arm=(np.random.choice(k,p=p_t))
        if optimal_arm==arm:
            optimal_action[i]=1
        reward_t=np.random.normal(loc=q_star[arm],scale=var)
        total_reward+=reward_t
        num_selections[arm]+=1
        R_avg=total_reward/(i+1)
        for j in range(k):
            if j==arm:
                H_t[j]=preferance_selected(j,H_t[j],alpha,reward_t,R_avg,p_t[j])
            else:
                H_t[j]=preferance_not_selected(j,H_t[j],alpha,reward_t,R_avg,p_t[j])
        for j in range(k):
            p_t[j]=softmax(j,H_t)
        avg_reward[i]=reward_t
    return avg_reward,optimal_action
```

```python
def one_sim_gb_wb(k,alpha,var,time_steps):        # gradient badit , 2000 simulations without baseline
    avg_reward=[0]*time_steps
    total_reward=0
    optimal_action=[0]*time_steps
    q_star=np.random.normal(loc=4,scale=1,size=(k))
    optimal_arm=np.argmax(q_star)
    opt_selections=0
    num_selections=[0]*k
    R_avg=0
    H_t=[0]*k
    p_t=[0.1]*k
    for i in range(time_steps):
        arm=int(np.random.choice(k,p=p_t))
        if optimal_arm==arm:
            optimal_action[i]=1
        reward_t=np.random.normal(loc=q_star[arm],scale=var)
        total_reward+=reward_t
        num_selections[arm]+=1
        for j in range(k):
            if j==arm:
                H_t[j]=preferance_selected(j,H_t[j],alpha,reward_t,R_avg,p_t[j])
            else:
                H_t[j]=preferance_not_selected(j,H_t[j],alpha,reward_t,R_avg,p_t[j])
        for j in range(k):
            p_t[j]=softmax(j,H_t)
        avg_reward[i]=reward_t
    return avg_reward,optimal_action
```

```python
def multi_sim_gb(k,alpha,var,exp,time_steps):        # with baseline, 2000 simulations
    avg_reward=[0]*time_steps
    optimal_action=[0]*time_steps
    for i in range(exp):
        a,b=one_sim_gb(k,alpha,var,time_steps)
        for j in range(time_steps):
            avg_reward[j]+=a[j]/exp
            optimal_action[j]+=(b[j]/exp)*100
    return avg_reward,optimal_action
```

```python
def multi_sim_gb_wb(k,alpha,var,exp,time_steps):        # without baseline 2000 simulations
    avg_reward=[0]*time_steps
    optimal_action=[0]*time_steps
    for i in range(exp):
        a,b=one_sim_gb_wb(k,alpha,var,time_steps)
        for j in range(time_steps):
            avg_reward[j]+=a[j]/exp
            optimal_action[j]+=(b[j]/exp)*100
    return avg_reward,optimal_action
```
