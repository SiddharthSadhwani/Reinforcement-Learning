## Question 2

```python
gamma_pi=np.zeros((25,25))  # coeffiecient matrix
r_pi=np.zeros(25) #constant matrix
r=5 #number of rows
c=5 #number of columns
gamma=0.9
A_initial=[0,1]
A_final=[4,1]
B_initial=[0,3]
B_final=[2,3]
A_reward=10
B_reward=5
pi=0.25 #probability of each action
x_steps=[1,-1,0,0]
y_steps=[0,0,1,-1]
```

```python
def next_cell(x,y): #returns a list of all valid cells we can transition to from (i,j)
    cells=[]
    for i in range(len(x_steps)):
        if (x-x_steps[i]>=0 and x-x_steps[i]<r) and (y-y_steps[i]>=0 and y-y_steps[i]<c):
            cells.append([x-x_steps[i],y-y_steps[i]])
    return cells
```

```python
for i in range(r): # for all s in S
    for j in range(c):# for all s in S
        pos=[] 
        pos.append(i)
        pos.append(j)
        next_states=np.zeros((r,c)) # for given s,  matrix of all possible s'
        if pos==A_initial: # if s=A
            next_states[A_final[0]][A_final[1]]=-(4*pi*gamma) # all four choices lead to A'
            next_states[i][j]=1 # no transition from s to s
            r_pi[i*r+j]=A_reward # total reward is calculated for all state s = pi*r for each action,r 
        elif pos==B_initial: # if s=B
            next_states[B_final[0]][B_final[1]]=-(4*pi*gamma)
            next_states[i][j]=1
            r_pi[i*r+j]=B_reward
        else: # if s belongs to states other than A and B
            cells=next_cell(i,j) # all valid moves
            if len(cells)==2: # corner case
                for k in cells:
                    next_states[k[0]][k[1]]=-(pi*gamma)  
                next_states[i][j]=(1-2*pi*gamma) # for transition from s to s and reward=-1
                r_pi[i*r+j]=2*-1*pi
            elif len(cells)==3: # edge case
                for k in cells:
                    next_states[k[0]][k[1]]=-(pi*gamma)
                next_states[i][j]=(1-pi*gamma)
                r_pi[i*r+j]=1*-1*pi
            elif len(cells)==4: #center case
                for k in cells:
                    next_states[k[0]][k[1]]=-(pi*gamma)
                next_states[i][j]=1
                r_pi[i*r+j]=0
        gamma_pi[i*r+j]=next_states.ravel()  # updating values in coefficient matrix          
```

```python
V_0=np.round(np.reshape(v_pi,(5,5)),1) # initialising V_0=V_pi so that it can be used in value iteration
```

```python
theta=0.000001 #setting theta value to capture convergence of V_pi
```

```python
while 1:
    delta=0
    for i in range(r):
        for j in range(c):
            v=V_0[i][j]
            pos=[]
            pos.append(i)
            pos.append(j)
            if pos==A_initial: # if s=A
                V_0[i][j]=A_reward+gamma*V_0[A_final[0]][A_final[1]] #updating V_pi
                delta=max(delta,abs(v-V_0[i][j])) #checking difference between old and new state values
            elif pos==B_initial: #if s=B
                V_0[i][j]=B_reward+gamma*V_0[B_final[0]][B_final[1]]
                delta=max(delta,abs(v-V_0[i][j]))
            else: # other cases where s is not equal to A or B
                cells=next_cell(i,j)
                if len(cells)==2:
                    value=[]
                    for k in cells:
                        value.append(0+gamma*V_0[k[0]][k[1]])
                    value.append(-1+gamma*V_0[i][j])
                    V_0[i][j]=max(value)
                    delta=max(delta,abs(v-V_0[i][j]))
                elif len(cells)==3:
                    value=[]
                    for k in cells:
                        value.append(0+gamma*V_0[k[0]][k[1]])
                    value.append(-1+gamma*V_0[i][j])
                    V_0[i][j]=max(value)
                    delta=max(delta,abs(v-V_0[i][j]))
                elif len(cells)==4:
                    value=[]
                    for k in cells:
                        value.append(0+gamma*V_0[k[0]][k[1]])
                    V_0[i][j]=max(value)
                    delta=max(delta,abs(v-V_0[i][j]))
    if delta<theta: # if the difference between the old and new state value less than theta then we stop the iteration
        break
```

## Policy Iteration

```python
r=4
c=4
gamma=1 # undiscounted case
terminal=[[0,0],[3,3]] # terminal cases
theta=0.000001
V_pi=np.zeros((r,c)) # state values for each state
pi_k=np.zeros((r,c,4)) # policy function for each state for each action
initial_pi=[0.25,0.25,0.25,0.25] #[left,right,up,down]
dirn=["left","right","up","down"]
```

```python
#initialization
for i in range(r):
    for j in range(c):
        pi_k[i][j]=initial_pi #intitializing with 0
```

```python
def pol_eval():
    print("Evaluating")
    while True:
        delta=0
        for i in range(r):
            for j in range(c):
                v=copy.deepcopy(V_pi[i][j]) # storing old state value
                pos=[]
                pos.append(i)
                pos.append(j)
                if pos not in terminal: # for non terminal states
                    V_pi[i][j]=0
                    if (i>=0 and i<r) and (j-1>=0 and j-1<c): # left valid move
                        V_pi[i][j]+=pi_k[i][j][0]*(-1+gamma*V_pi[i][j-1])
                    else: # else left move is out of bounds
                        V_pi[i][j]+=pi_k[i][j][0]*(-1+gamma*v)
                    if (i>=0 and i<r) and (j+1>=0 and j+1<c): # same for right
                        V_pi[i][j]+=pi_k[i][j][1]*(-1+gamma*V_pi[i][j+1])
                    else:
                        V_pi[i][j]+=pi_k[i][j][1]*(-1+gamma*v)
                    if (i-1>=0 and i-1<r) and (j>=0 and j<c): # same for up
                        V_pi[i][j]+=pi_k[i][j][2]*(-1+gamma*V_pi[i-1][j])
                    else:
                        V_pi[i][j]+=pi_k[i][j][2]*(-1+gamma*v)
                    if (i+1>=0 and i+1<r) and (j>=0 and j<c): # same for down
                        V_pi[i][j]+=pi_k[i][j][3]*(-1+gamma*V_pi[i+1][j])
                    else:
                        V_pi[i][j]+=pi_k[i][j][3]*(-1+gamma*v)
                    delta=max(delta,abs(v-V_pi[i][j])) # computing difference between old and new state values
#                     print(delta)
        if delta<theta: # if difference less than  theta bound, then we stop evaluation as we have converged to V_pi
            break
```

```python
while True:
    print("ITERATION\n")
    pol_eval() # policy evaluation
    print(V_pi) 
    print("Improving\n")
    state=True # state true to capture condition of optimal policy
    for i in range(r): # for each state s in S
        for j in range(c):
            old=copy.deepcopy(pi_k[i][j]) # saving old policy function
#             print("old= ",old)
            pos=[]
            pos.append(i)
            pos.append(j)
            if pos not in terminal:
                value=[0]*len(dirn)
                if (i>=0 and i<r) and (j-1>=0 and j-1<c): # calculating r + gamma*Vs' for left
                    value[0]=(-1+gamma*V_pi[i][j-1])
                else:
                    value[0]=(-1+gamma*V_pi[i][j])
                if (i>=0 and i<r) and (j+1>=0 and j+1<c): # same for right move
                    value[1]=(-1+gamma*V_pi[i][j+1])
                else:
                    value[1]=(-1+gamma*V_pi[i][j])
                if (i-1>=0 and i-1<r) and (j>=0 and j<c): # same for up move
                    value[2]=(-1+gamma*V_pi[i-1][j])
                else:
                    value[2]=(-1+gamma*V_pi[i][j])
                if (i+1>=0 and i+1<r) and (j>=0 and j<c): # same for down move
                    value[3]=(-1+gamma*V_pi[i+1][j])
                else:
                    value[3]=(-1+gamma*V_pi[i][j])
#                 print("value= ",value)
                best=np.argwhere(value == np.amax(value)).flatten().tolist() # getting the optimal moves from all possible moves
#                 print("best= ",best)
                new=[0]*len(dirn)
                for b in range(len(new)): # iterating over all optimal moves. This takes care of the bug mentioned in the exercise 4.4 as now we change probabilites to go to all optimal states rather than only the first occurence
                    if b in best:
                        new[b]=1/len(best) # 1/total optimal actions
                pi_k[i][j]=new # assigning new policy function
#                 print("new= ",pi_k[i][j])
                st=0
                for g in range(len(old)): # if old and new policy function same then that means we have converged to the optimal policy( should be for all states)
                    if old[g]!=pi_k[i][j][g]:
                        state=False
                        break
                print(pi_k[i][j])
#                 print(old," ",new)
#                 print("hello")
    if state: # if converged then returning 
        final_directions=[]
        print("FINAL\n")
        print("V*\n")
        print(V_pi)
        print("pi*\n")
        print(pi_k)
        for i in range(r):
            for j in range(c):
                pos=[]
                pos.append(i)
                pos.append(j)
                if pos in terminal: # if at terminal state
                    final_directions.append("Terminal")
                else: # if at any other state
                    act=np.argwhere(pi_k[i][j] == np.amax(pi_k[i][j])).flatten().tolist() # getting the optimal actions
                    s=""
                    for y in act:
                        s+=dirn[y]+" "
                    final_directions.append(s)
        print("Optimal Actions\n")
        print(final_directions) # all optimal actions for each state
        break
```
