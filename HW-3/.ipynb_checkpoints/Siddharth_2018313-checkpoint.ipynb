```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
Cards={}
Cards['A']=11
Cards['J']=10
Cards['Q']=10
Cards['K']=10
```

```python
for i in range(2,10):
    Cards[str(i)]=i
```

```python
def draw_player():
    card_1=np.random.choice(list(Cards.keys()))
    card_2=np.random.choice(list(Cards.keys()))
    p_sum=0
    use_ace=0
    if card_1=='A':
        use_ace=1
    p_sum+=Cards[card_1]
    if card_2=='A' and p_sum<=10:
        use_ace=1
        p_sum+=Cards[card_2]
    elif card_2=='A':
        p_sum+=1
    else:
        p_sum+=Cards[card_2]
    return p_sum,card_1,card_2,use_ace
```

```python
def BlackJack():
    player_sum=0
    dealer_sum=0
    face_up=0
    face_down=0
    use_ace=0
    variables=[]
    
    player_sum,p_card1,p_card2,use_ace=draw_player()
    face_up=np.random.choice(list(Cards.keys()))
    face_down=np.random.choice(list(Cards.keys()))
    dealer_sum=Cards[face_up]+Cards[face_down]
    
    variables.append([player_sum,Cards[face_up],use_ace])
    
    if player_sum==21:
        if dealer_sum==21:
            return variables,0
        else:
            return variables,1
    elif dealer_sum==21:
        return variables,-1
    while True:
        pick_card=np.random.choice(list(Cards.keys()))
        if pick_card=='A':
            if player_sum<=10:
                total=player_sum+Cards[pick_card]
                use_ace=1
            else:
                total=player_sum+1
        else:
            total=player_sum+Cards[pick_card]
        if total>21:
            if use_ace==1:
                total=player_sum-10
                use_ace=0
            else:
                player_sum=total
                variables.append([player_sum,Cards[face_up],use_ace])
                return variables,-1
        if total==21 or total==20:
            player_sum=total
            variables.append([player_sum,Cards[face_up],use_ace])
            break
        player_sum=total
        variables.append([player_sum,Cards[face_up],use_ace])
    while(True):
        deal_card=np.random.choice(list(Cards.keys()))
        dealer_sum+=Cards[deal_card]
        if dealer_sum>21:
            return variables,1
        elif dealer_sum>=17:
            break
    if player_sum==dealer_sum:
        return variables,0
    elif player_sum>dealer_sum:
        return variables,1
    else:
        return variables,-1
```

```python
def MC(episodes):
    counts=np.ones((10,10,2))
    values=np.zeros((10,10,2))
    for i in tqdm(range(episodes)):
        state,reward=BlackJack()
        for st in range(len(state)-1,-1,-1):
            if state[st][0]<=21:
                a=state[st][0]
                a=a-12
                if state[st][1]==11:
                    b=0
                else:
                    b=state[st][1]
                    b=b-1
                c=state[st][2]
                if state[st] not in state[:st-1]:
                    values[a,b,c]=(values[a,b,c]*counts[a,b,c]+reward)/(counts[a,b,c]+1)
                    counts[a,b,c]+=1
    return counts,values
```

```python
def BlackJack(policy):
    player_sum=0
    dealer_sum=0
    face_up=0
    face_down=0
    use_ace=0
    variables=[]
    
    player_sum,p_card1,p_card2,use_ace=draw_player()
    face_up=np.random.choice(list(Cards.keys()))
    face_down=np.random.choice(list(Cards.keys()))
    dealer_sum=Cards[face_up]+Cards[face_down]
    
    
    if player_sum==21:
        if dealer_sum==21:
            return variables,0
        else:
            return variables,1
    elif dealer_sum==21:
        return variables,-1
    else:
        while True:
            vals=0
            if Cards[face_up]==11:
                vals=1
            else:
                vals=Cards[face_up]
            if(policy[player_sum-12][vals-1][use_ace]==1):
                variables.append([player_sum,Cards[face_up],use_ace,1])
                pick_card=np.random.choice(list(Cards.keys()))
                player_sum+=Cards[pick_card]

                if player_sum>21:
                    if use_ace==1:
                        player_sum-=10
                        use_ace=0
                if player_sum>21:
                    return variables,-1
                elif player_sum==21:
                    return variables,1
            else:
                variables.append([player_sum,Cards[face_up],use_ace,0])
                break
        use_ace=0
        while(True):
            deal_card=np.random.choice(list(Cards.keys()))
            if deal_card=='A' and dealer_sum<=10:
                dealer_sum+=Cards[deal_card]
                use_ace=1
            elif deal_card=='A':
                dealer_sum+=1
            else:
                dealer_sum+=Cards[deal_card]
            if dealer_sum==21:
                return variables,-1
            elif dealer_sum>=22:
                if use_ace==1:
                    dealer_sum-=10
                    use_ace=0
                else:
                    return variables,1
            elif(dealer_sum>16):
                break
        if player_sum==dealer_sum:
            return variables,0
        elif player_sum>dealer_sum:
            return variables,1
        else:
            return variables,-1
```

```python
def MCES(episodes):
    counts=np.ones((10,10,2,2))
    state_action=np.zeros((10,10,2,2))
    policy=np.ones((10,10,2))
    policy[8:,:,:]=0
    for i in tqdm(range(episodes)):
        state,reward=BlackJack(policy)
        for st in range(len(state)-1,-1,-1):
            if state[st][0]<=21:
                a=state[st][0]
                a=a-12
                if state[st][1]==11:
                    b=0
                else:
                    b=state[st][1]
                    b=b-1
                c=state[st][2]
                act=state[st][3]
                if state[st] not in state[:st-1]:
                    state_action[a,b,c,act]=(state_action[a,b,c,act]*counts[a,b,c,act]+reward)/(counts[a,b,c,act]+1)
                    counts[a,b,c,act]+=1
                policy[a,b,c]=np.argmax(state_action[a,b,c,:])
    return state_action,policy
```

```python
a,b=MCES(500000)
opti_v=np.max(a,axis=3)
```

```python
Actions={'L':0,'R':1}
true_val=[0,1/6,2/6,3/6,4/6,5/6,0]
Names=['A','B','C','D','E']
```

```python
def td0(alpha):
    V=np.zeros(7)
    V[1:6]=0.5
    episodes=100
    mser=np.zeros(episodes)
    for ep in range(episodes):
        if(alpha==0.1 and ep==0):
            plt.plot(Names,V[1:6],label=ep)
        cur_state=len(true_val)//2
        while(True):
            action=random.randint(0,1)
            if action==0:
                next_state=cur_state-1
            else:
                next_state=cur_state+1
            if next_state==6:
                reward=1
            else:
                reward=0
            V[cur_state]+=alpha*(V[next_state]+reward-V[cur_state])
            cur_state=next_state
            if cur_state==6 or cur_state==0:
                break
        if ep==1 or ep==10 or ep==99:
            if ep==99:
                e=ep+1
            else:
                e=ep
            plt.plot(Names,V[1:6],label=e)
        mser[ep]=np.sqrt(np.mean(np.power(V-true_val,2)))
    if(alpha==0.1):
        plt.plot(Names,true_val[1:6],label='True')
        plt.legend()
        plt.show()
    return mser
```

```python
def td0_wp(alpha):
    V=np.zeros(7)
    V[1:6]=0.5
    episodes=100
    mser=np.zeros(episodes)
    for ep in range(episodes):
        cur_state=len(true_val)//2
        while(True):
            action=random.randint(0,1)
            if action==0:
                next_state=cur_state-1
            else:
                next_state=cur_state+1
            if next_state==6:
                reward=1
            else:
                reward=0
            V[cur_state]+=alpha*(V[next_state]+reward-V[cur_state])
            cur_state=next_state
            if cur_state==6 or cur_state==0:
                break
        mser[ep]=np.sqrt(np.mean(np.power(V-true_val,2)))
    return mser
```

```python
def monty(alpha):
    V=np.zeros(7)
    V[1:6]=0.5
    episodes=100
    mser=np.zeros(episodes)
    for ep in range(episodes):
        cur_state=len(true_val)//2
        s_a=[]
        while(True):
            action=random.randint(0,1)
            if action==0:
                next_state=cur_state-1
            else:
                next_state=cur_state+1
            if next_state==6:
                reward=1
            else:
                reward=0
            s_a.append([cur_state,reward,action])
            cur_state=next_state
            if cur_state==6 or cur_state==0:
                break
        G=0
        s_a.reverse()
        for i in s_a:
            G+=i[1]
            V[i[0]]+=alpha*(G-V[i[0]])
        mser[ep]=np.sqrt(np.mean(np.power(V-true_val,2)))
    return mser                  
```

```python
mser_td=np.zeros((100,100))
runs=100
for r in range(runs):
    mser_td[r]=td0_wp(0.1)
pt=np.mean(mser_td,axis=0)
plt.plot(pt,label="TD , a=0.1")

mser_td=np.zeros((100,100))
runs=100
for r in range(runs):
    mser_td[r]=td0_wp(0.15)
pt=np.mean(mser_td,axis=0)
plt.plot(pt,label="TD , a=0.15")

mser_td=np.zeros((100,100))
runs=100
for r in range(runs):
    mser_td[r]=td0_wp(0.05)
pt=np.mean(mser_td,axis=0)
plt.plot(pt,label="TD , a=0.05")

mser_mc=np.zeros((100,100))
runs=100
for r in range(runs):
    mser_mc[r]=monty(0.01)
pt=np.mean(mser_mc,axis=0)
plt.plot(pt,label="MC , a=0.01")

mser_mc=np.zeros((100,100))
runs=100
for r in range(runs):
    mser_mc[r]=monty(0.02)
pt=np.mean(mser_mc,axis=0)
plt.plot(pt,label="MC , a=0.02")

mser_mc=np.zeros((100,100))
runs=100
for r in range(runs):
    mser_mc[r]=monty(0.03)
pt=np.mean(mser_mc,axis=0)
plt.plot(pt,label="MC , a=0.03")

mser_mc=np.zeros((100,100))
runs=100
for r in range(runs):
    mser_mc[r]=monty(0.04)
pt=np.mean(mser_mc,axis=0)
plt.plot(pt,label="MC , a=0.04")

plt.ylabel("Empirical RMS error, averaged over states")
plt.xlabel("Walks/ Episodes")
plt.legend()
plt.show()
```
